```python
import pandas as pd
import numpy as np
```

```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
rice_cammeo_and_osmancik = fetch_ucirepo(id=545) 
  
# data (as pandas dataframes) 
X = rice_cammeo_and_osmancik.data.features 
y = rice_cammeo_and_osmancik.data.targets 
  


```

```python
data = pd.concat([X, y], axis = 1)
data.shape
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.to_csv('./data/rice_data_clean.csv',index=False)
data.shape
```

```python
unseen_data = data.sample(frac=.20, random_state=42)
df = data.drop(unseen_data.index)
unseen_data
```

```python
unseen_data.to_csv('./data/unseen_data.csv',index=False)
df.to_csv('./data/rice_main.csv', index=False)
unseen_data.shape, df.shape
```
